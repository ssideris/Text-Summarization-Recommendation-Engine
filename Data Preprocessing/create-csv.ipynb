In this module we will load multiple .json files from multiple directories and convert them to a .csv format.

```python
#import libraries
import json
import gzip
import os
import pandas as pd
```

```python
#We will create a function that will read multiple .json files from a given directory 
#and transform them to csv format (dataframe)
def readData(input_path,split_type,cpc_code):
    #assign all the filenames in the directory
    file_names = os.listdir(os.path.join(input_path,split_type,cpc_code))
    #iterate through all the filenames using a for-loop
    for i in range(0,len(file_names)):
        # reading one of the gz files.
        file_name = file_names[i]
        
        #we open the gzip file to iterate through the json files inside it
        with gzip.open(os.path.join(input_path,split_type,cpc_code,file_name),'r') as fin:
            for row in fin:
                #we load the json files
                json_obj = json.loads(row)
                
                #we append the information we need in the dataframe
                mydata.loc[len(mydata)] = [json_obj["publication_number"], json_obj["description"], json_obj["abstract"]]

                break
            
    return mydata
```

```python
#We create an empty dataframe
mydata = pd.DataFrame(columns = ['Publication_Number', 'Description', 'Summary'])

#We iterate through the different directories that include the json files with a for loop
for k in ["train", "test", "val"]:
    for l in ["a", "b", "c", "d", "e", "f", "g", "h", "y"]:
        data = readData("/bigPatentData", k, l)
        mydata.append(data)
mydata
```

```python
#remove row 1838 because it had a length greater than the length limit for each dataframe's cell
mydata = mydata.drop(1837)
mydata
```

```python
#We export the dataframe to csv format
mydata.to_csv('initial_dataset.csv')
```
